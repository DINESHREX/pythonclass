```python
file = open("sample.txt", "r")
content = file.read()
content = content.replace(","," ").replace("."," ").lower()
words = content.split()
freq ={}
for word in words:
    if word in freq:
        freq[word] +=1
    else:
        freq[word]=1
print("Word Frequency")
for f in freq:
    print(f,":", freq[f])
```
