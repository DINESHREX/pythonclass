```python
s=input("Enter the Skill :").lower()
e=int(input("Enter the Experience (if freshers enter 0) :"))
l=input("Enter the Location :").lower()

if s=="python" and e >= 1 and l=="chennai":
    print("You are eligible for Infosys")
elif s=="python" and e == 0 and l=="coimbatore":
    print("You are eligible for Wipro")
elif s=="java" and e == 0 and l=="chennai":
    print("You are eligible for TCS") 
elif s=="java" and e >= 1 and l=="bangalore":
    print("You are eligible for Cognizant")
else:
    print("There are no role match for your  credentials")
```
