```python
num1=int(input("Enter The First No: "))
num2=int(input("Enter The Second No: "))
op= input("Enter The Operation (+,-,*,//,%): ")

if op == "+":
 result = num1 + num2
 print(result)
elif op == "-":
 result = num1 - num2
 print(result)
elif op == "*":
 result = num1 * num2
 print(result)
elif op == "//":
 result = num1 // num2
 print(result)
elif op == "%": 
 result = num1 % num2
 print(result)
else:
 print("invalid operator or numbers")
    
```

```python
num1 = int(input("Enter The First No: "))
num2 = int(input("Enter The Second No: "))
op = input("Enter The Operation (+,-,*,//,%): ")

operations = {
    "+": num1 + num2,
    "-": num1 - num2,
    "*": num1 * num2,
    "//": num1 // num2 if num2 != 0 else "Cannot divide by zero",
    "%": num1 % num2 if num2 != 0 else "Cannot divide by zero"
}

result = operations.get(op, "Invalid operator")

print(result)
```
