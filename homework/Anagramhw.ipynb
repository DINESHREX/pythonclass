```python
def anagram(str1, str2):
    chcount = {}
    i = 0

    while True:

        if str1[i:i+1] == "" and str2[i:i+1] == "":
            break

        if str1[i:i+1] == "" or str2[i:i+1] == "":
            return False

        if str1[i] in chcount:
            chcount[str1[i]] += 1
        else:
            chcount[str1[i]] = 1

        if str2[i] in chcount:
            chcount[str2[i]] -= 1
        else:
            chcount[str2[i]] = -1

        i += 1

    for v in chcount:
        if chcount[v] != 0:
            return False

    return True


str1 = input("Enter The Word ")
str2 = input("Enter The Word ")

if anagram(str1, str2):
    print("Anagram")
else:
    print("Not Anagram")
```
