```python
# only str - bool 
st="Dinesh"
stb = bool(st)  
print(type(stb))
```

```python
# str to int 
st = "123" # valid 
sti = int(st)
print(type(sti))
```

```python
# str to float 
st = "12.3" # valid 
sti = float(st)
print(type(sti))
```

```python
# str to complex 
st = "12" # valid 
sti = complex(st)
print(type(sti))
```
