```python
#Arithmetic Operators
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
```

```python
#Assignment Operators

x = 10
x += 5
print(x)
```

```python
s = 10
s *= 2
print(s)
```

```python
d=10
d **= 2
print(d)
```

```python
#Comparison Operator

p = 10
q = 20
print(p == q)
print(p != q)
print(p > q)
print(p < q)
print(p >= 10)
print(q <= 15)
```
