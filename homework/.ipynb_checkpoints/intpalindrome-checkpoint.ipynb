```python
num = int(input("Enter The Number : "))
if num <0:
     print(num ,"Not a Palindrome")
else:
    original =num
    reverse=0
    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit\
        num // 10
        if reverse == original:
            print(num ,"is Palindrome")
        else:
            print(num ,"Not a Palindrome")
```
