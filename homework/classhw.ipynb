```python
class University:
    def uni_info(self):
        print("This is a University")

class College(University):
    def college_info(self):
        print("This is a College")

class Department(College):
    def dept_info(self):
        print("This is a Department")

d = Department()

d.uni_info()
d.college_info()
d.dept_info()
```

```python
class University:
    def uni_info(self):
        print("University Level")

class College(University):
    def college_info(self):
        print("College Level")

class Department(College):
    def dept_info(self):
        print("Department Level")

d = Department()
d.uni_info()
d.college_info()
d.dept_info()
```

```python
class University:
    def uni_info(self):
        print("University Information")

class Department:
    def dept_info(self):
        print("Department Information")

class College(University, Department):
    def college_info(self):
        print("College Information")

c = College()
c.uni_info()
c.dept_info()
c.college_info()
```
