```python
x=10 
b=11
class calc:
    def add(calc):
        c=x+b
        return c
    def sub(calc):
        c=x-b
        return c
    def mul(calc):
        c=x*b
        return c
cal = calc()
k = cal.add()
s = cal.sub()
m = cal.mul()
print(m)
```

```python
class obj():
    def __init__(self,a,b,c):
        self.num1 =a
        self.num2 =b
        self.num3 =c
    def add(self):
        print(self.num1+self.num2)
    def sub(self):
        print(self.num3-self.num2)

k=obj(3,4,5)
k.add()
k.sub()
    
```
