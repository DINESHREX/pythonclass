```python
# check anagram

str1 =input("Enter a word")
str2 =input("Enter a word")
cs1 =0
cs2 =0

for i in str1:
    cs1 +=1
for i in str2:
    cs2 +=1
if cs1 != cs2:
    print("Not a Anagram")
else:
    count =0
    for i in str1:
        for j in str2:
            if i == j:
                count+=1
                break
    if count == cs1:
        print("Its an Anagram")
    else:
        print("Not an anagram")
        
```

```python
# Ascending Order 
n= [12, 45, 7, 23, 56, 89, 34]
ct = 0
for i in n:
    ct+=1
for i in range(ct):
    for j in range(i+1, ct):
        if n[i]>n[j]:
            temp =n[i]
            n[i]=n[j]
            n[j] = temp
print('Ascending Order', n)
```

```python
# Descending Order
n= [12, 45, 7, 23, 56, 89, 34]
ct = 0
for i in n:
    ct+=1
for i in range(ct):
    for j in range(i+1,ct):
        if n[i]<n[j]:
            temp =n[j]
            n[j] = n[i]
            n[i] = temp

print('Descending Order', n)       
```

```python
# Merge and Sort the List
list1 = [12, 45, 7]
list2 = [23, 56, 34]
l_t=[]

for i in list1:
    l_t.append(i)
for i in list2:
    l_t.append(i)
# print(l_t)

count =0
for i in l_t:
    count+=1

for i in range(count):
    for j in range(i+1 , count):
            if l_t[i] > l_t[j]:
                temp = l_t[i]
                l_t[i] = l_t[j]
                l_t[j] = temp
print(f"The Merged Sort {l_t}")
```
