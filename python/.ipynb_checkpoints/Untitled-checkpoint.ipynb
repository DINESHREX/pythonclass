```python
# Anagram
txt1 = input("Enter a text")
txt2 = input("Enter a text")
t1 =0
t2 =0

for i in txt1:
    t1+=1
for i in txt2:
    t2+=1

if t1 != t2:
    print("Not Anagram")
else:
    count =0
    for i in txt1:
        for j in txt2:
            if j ==i:
                count +=1
    if count == t1:
        print("Its Anagram")
            
    
```

```python
# palindrome
txt = input("Enter text: ")
rev =""
for i in txt:
    rev = i+rev
if txt == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# prime or not prime

num = int(input("Enter The number: "))
count = 0

for i in range(1, num+1):
    result = num % i
    if result ==0:
        count +=1
if count == 2:
    print("Its Prime")
else:
    print("Not Prime")
```

```python
# fibonacci
num = int(input("Enter The Number: "))
a=0
b=1
for i in range(num+1):
    print(a, end=" ")
    c=a+b
    a=b
    b=c
    
```

```python
# max and min num
numbers=[1,5,89,34,222,45,678,23]

max_num = numbers[0]
min_num = numbers[0]

for i in numbers:
    if i > max_num:
        max_num = i
    if i < min_num:
        min_num =i
print("Maximum : ", max_num)
print("Mimimum : ", min_num)
```

```python
# max and min num
numbers=[1,5,89,34,222,45,678,23]

max_num = numbers[0]
min_num = numbers[0]

for i in numbers:
    if i > max_num:
        max_num = i
    if i < min_num:
        min_num =i

s_max_num = min_num
s_min_num = max_num

for i in numbers:
    if i != max_num and i > s_max_num:
        s_max_num = i
    if i != min_num  and i < s_min_num:
        s_min_num = i
    else:
        i+=1
print("Maximum : ", s_max_num)
print("Mimimum : ", s_min_num)
```

```python
# ascending
num=[1,5,89,34,222,45,678,23]
ct=0
for i in num:
    ct+=1

numa= num

for i in range(ct):
    for j in range(i+1,ct):
        if numa[i]>numa[j]:
            temp = numa[i]
            numa[i]=numa[j]
            numa[j]=temp

print(numa)
```

```python
# descending
num=[1,5,89,34,222,45,678,23]
ct=0
for i in num:
    ct+=1
numd=num

for i in range(ct):
    for j in range(i+1,ct):
        if numd[i]<numd[j]:
            temp=numd[j]
            numd[j]=numd[i]
            numd[i]=temp
print(numd)
```

```python
# to check vowels and consonant

txt = input("Enter the Text: ").lower()
vowel =0
consonant =0
for i in txt:
    if i.isalpha():
        if i in "aeiou":
            vowel +=1
        else:
            consonant+=1
print("Vowels", vowel)
print("Consonant", consonant)
```

```python
# frequency of num

numbers = [1, 2, 2, 3, 1, 4, 2]
ls =[]
for i in numbers:
    if i not in ls:
        count =0
        ls.append(i)
        for j in numbers:
            if j ==i:
                count+=1
        print(i,":",count)  
```

```python
# amstrong number

num = int(input("Enter The Number : "))
original = num
temp = num
count =0

while temp > 0:
    count +=1
    temp = temp // 10

total =0
power = count
temp = num
while temp>0:
    digit = temp % 10
    total += digit ** power
    temp = temp // 10
if total == original:
    print("Amstrong")
else:
    print("Not Amstrong")
```

```python
# remove duplicates from a list

nums = [12, 45, 7, 23, 56, 89, 34, 23, 45, 12]
n_list=[]
for i in nums:
    if i not in n_list:
        n_list.append(i)
print(n_list)
```

```python
# merge and sort lists

list1 = [12, 45, 7]
list2 = [23, 56, 34]
l_t=[]
for i in list1:
    l_t.append(i)
for i in list2:
    l_t.append(i)
    
count = 0
for i in l_t:
    count+=1

for i in range(count):
    for j in range(i+1,count):
        if l_t[i] > l_t[j]:
            temp =l_t[i]
            l_t[i] =l_t[j]
            l_t[j] =temp
print(l_t)
```
