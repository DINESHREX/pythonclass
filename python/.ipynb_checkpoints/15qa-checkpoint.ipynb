```python
text1 =input("Enter The Text")
text2 =input("Enter The Text")
t1 =0
t2 =0
for i in text1:
    t1+=1
for i in text2:
    t2+=1

if t1 != t2:
    print("Not Anagram")
else:
    count =0
    for i in text1:
        for j in text2:
            if i == j:
                count +=1
    if t1 == count:
        print("Its Anagram")
    else:
        print("Not Anagram")
```

```python
text =input("Enter The word ")
rev=""
for i in text:
    rev = i+rev
if text == rev:
    print('Palindrome')
else:
    print("Not Palindrome")
```

```python
n=int(input("Enter The Number:"))
count =0
for i in range(1,n+1):
    if n % i == 0:
        count +=1
if count == 2:
    print("Its Prime")
else:
    print("Not Prime")
```

```python
num = [12, 45, 7, 23, 56, 89, 34]
max_num=num[0]
min_num=num[0]
for i in num:
    if i > max_num:
        max_num = i
    if i< min_num:
        min_num = i
print("Maximum", max_num)
print("Minimum", min_num)
```

```python
num = [12, 45, 7, 23, 56, 89, 34]
max_num=num[0]
min_num=num[0]
for i in num:
    if i > max_num:
        max_num = i
    if i< min_num:
        min_num = i
s_largest =num[0]
s_smallest =num[0]
for i in num:
    if i != max_num and s_largest < i:
        s_largest = i
    if i != min_num and s_smallest > i:
        s_smallest = i
print("Second Largest",s_largest)
print("Second Smallest",s_smallest)
```

```python
num = [12, 45, 7, 23, 56, 89, 34]
ct = 0
for i in num:
    ct+=1
for i in range(ct):
    for j in range(i+1,ct):
        if num[i] > num[j]:
            temp=num[i]
            num[i]=num[j]
            num[j]=temp
print("Ascending Order", num)

for i in range(ct):
    for j in range(i+1,ct):
        if num[i] < num[j]:
            temp=num[j]
            num[j]=num[i]
            num[i]=temp
print("Descending Order", num)
```

```python
text = input("Enter The text:")
vowel =0
consonant=0
for i in text:
    if i.isalpha():
        if i in "aeiou":
            vowel +=1
        else:
            consonant+=1
print("Vowels", vowel)
print("Consonant", consonant)
```

```python
num = int(input("Enter A number"))
sum =0
while num >0:
    digit = num%10
    sum += digit
    num = num // 10
print("Sum", sum)
```
