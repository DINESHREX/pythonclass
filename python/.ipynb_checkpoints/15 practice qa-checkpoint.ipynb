```python
# fibonacci

n=int(input("Enter The Number: "))
a=0
b=1
for i in range(n):
    print(a, end=" ")
    c=a+b
    a=b
    b=c
```

```python
# min max with loop 
numbers=[1,5,89,34,222,45,678,23]
max_num=numbers[0]
min_num=numbers[0]
for i in numbers:
    if i>max_num:
        max_num = i
    if i<min_num:
        min_num = i
print("Maximum :" ,max_num)
print("Minimum :" ,min_num)
    
```

```python
# sum of digits

num =int(input("Enter a Number:"))
total =0
while num>0:
    digit = num % 10
    total += digit
    num = num//10
print("Sum Of Digits:" ,total)
```

```python
# count vowels consonant

text=input("Enter The Text : ").lower()
vowels =0
consonants =0
for i in text:
    if i.isalpha():
        if i in "aeiou":
            vowels+=1
        else:
            consonants+=1
print("Vowels",vowels )
print("Consonants",consonants )
            
```

```python
# factorial
n =int(input("Enter The Range Of Number: "))
fact =1
for i in range(1,n+1):
    fact=fact*i
    print(fact, end=" ")
    
```

```python
# odd or even
n =int(input("Enter The Number: "))
if n % 2 ==0:
    print(n ,"is Even")
else:
    print(n ,"is Odd")
```

```python
# largest of three
a =int(input("Enter The First Number: "))
b =int(input("Enter The Second Number: "))
c =int(input("Enter The Third Number: "))
largest=a
if b > largest:
    largest =b
if c > largest:
    largest =c
print(largest,"is Largest ")
```

```python
# frequency of the element

numbers = [1, 2, 2, 3, 1, 4, 2]
for i in numbers:
    count = 0
    for j in numbers:
        if j==i:
            count+=1
    print(i,":",count)
        
```
